**4.1 Data Preprocessing**

*   In this dataset, the original columns lacked descriptive categories for gender and age groups, which I created to improve clarity for my analysis. *   First, because the original Sex column (Column B) encoded data numerically as 0 and 1, I inserted an adjacent column named Gender and applied the Excel formula =IF(B2=1,"Male","Female") to clearly convert 1 into "Male" and 0 into "Female".
*   Second, to simplify individual numerical ages into broader, more meaningful lifespans, I added an Age Group column next to the Age column (Column A). Using a nested Excel conditional statement, =IF(A2<=34,"Young",IF(A2<=50,"Middle",IF(A2<=64,"Senior","Elderly"))), I successfully binned the continuous ages into four specific macro-categories: Young (ages 18–34), Middle (ages 35–50), Senior (ages 51–64), and Elderly (ages 65 and older).

**Plotting graphs, charts, and plots.**

Question1: What is the age distribution of the patients in this study? Is it skewed toward older or younger populations?

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=heart_disease_df, x='Age Group',bins=20,kde=True,color='g')
plt.title('Age Distribution of Patients')
plt.xlabel('Age Group')
plt.ylabel('Frequency')

```

The chart shows that the dataset is heavily skewed toward older populations, with the vast majority of patients falling into the Senior and Middle age categories. In contrast, there are very few elderly patients and almost no young patients included in the study. Therefore, the data primarily reflects individuals in their mid-career to later life stages.

Question2: What is the gender breakdown of the patients? Are males or females more represented?

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=heart_disease_df['Gender'].value_counts().index,y=heart_disease_df['Gender'].value_counts())
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Frequency')
```

Based on the barplot, there are significantly more male patients in the dataset than female patients. Therefore, males are much more represented, making up the vast majority of the total individuals studied.

Question3: Do patients diagnosed with heart disease tend to have significantly higher cholesterol levels compared to those who don't?

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Heart Disease',y='Cholesterol',hue='Gender',data=heart_disease_df)
plt.title('Cholesterol Distribution by Heart Disease')
```

The chart reveals that patients with heart disease ("Presence") generally have noticeably higher cholesterol levels compared to those without it ("Absence"). Additionally, regardless of heart disease status, female patients consistently display higher average cholesterol levels than male patients.

Question3: Does the maximum heart rate achieved drop as a patient ages, and does this trend differ based on whether they have heart disease?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Age',y='Max HR',hue='Heart Disease',data=heart_disease_df)
plt.title('Age vs. Maximum heart rate')
```

The chart confirms that maximum heart rate generally drops as patients get older, showing a downward trend for both groups from left to right. However, this trend differs because patients with heart disease ("Presence", the blue line) consistently maintain a lower maximum heart rate across almost all age groups compared to healthy patients ("Absence", the orange line).

Question4: How does the relationship between Age and Blood Pressure change when comparing men versus women?

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age',y='BP',hue='Gender',data=heart_disease_df)
plt.title('Blood Pressure vs. Age')
```

the scatterplot shows a general upward trend, indicating that blood pressure tends to increase as patients get older. This rising pattern is quite similar for both men and women, with both genders widely and evenly mixed across the age groups.

Question5: Which type of chest pain is most commonly reported by patients who are ultimately diagnosed with heart disease?

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Chest pain type',hue='Heart Disease',data=heart_disease_df)
plt.title('Chest Pain Types by Heart Disease')
```

The countplot shows shows that chest pain type 4 is by far the most commonly reported type among patients diagnosed with heart disease ("Presence"). In contrast, patients who do not have heart disease ("Absence") are most frequently associated with chest pain type 3.

Question6: Does having a high fasting blood sugar level significantly correlate with a positive heart disease diagnosis?

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='FBS over 120',hue='Heart Disease',data=heart_disease_df)
plt.title('Fasting blood sugar level by Heart Disease')
```

The countplot shows that a high fasting blood sugar level (value 1) does not strongly correlate with a positive heart disease diagnosis, as the number of patients remains low and evenly split between presence and absence. Instead, the vast majority of patients in both groups actually have lower fasting blood sugar levels (value 0).

Question7: How do the results of an electrocardiographic test differ between patients with and without heart disease?

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='EKG results',hue='Heart Disease',data=heart_disease_df)
plt.title('Electrocardiographic test by Heart Disease')
```

The countplot shows that patients without heart disease ("Absence") are most likely to have an EKG result of 0. On the other hand, patients diagnosed with heart disease ("Presence") are significantly more likely to have an EKG result of 2, making up the majority of that category.
